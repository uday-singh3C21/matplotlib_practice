# histograph

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age= np.random.randint(18,45,(100))
py_student_age=np.random.randint(19,45,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
plt.show()
```

```python
plt.hist(ml_student_age,rwidth=0.9)
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist(ml_student_age,bins,rwidth=0.9,orientation='horizontal')
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist(ml_student_age,bins,rwidth=0.9,orientation='horizontal',color='b')
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist(ml_student_age,bins,rwidth=0.9,orientation='horizontal',color='b',label='ml_student')
plt.legend()
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist(ml_student_age,bins,rwidth=0.8,orientation='horizontal',color='b',label='ml_student')
plt.hist(py_student_age,bins,rwidth=0.8,orientation='vertical',color='y',label='ml_student')
plt.legend()
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist(ml_student_age,bins,rwidth=0.8,orientation='vertical',color='b',label='ml_student')
plt.hist(py_student_age,bins,rwidth=0.8,orientation='vertical',color='y',label='py_student')
plt.legend()
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,orientation='vertical',color=['b','r'],label=['ml_student','py_student'])
#plt.hist(py_student_age,bins,rwidth=0.8,orientation='vertical',color='y',label='py_student')
plt.legend()
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
bins=[15,20,25,30,35,40,45]
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.figure(facecolor='lightgrey')
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,orientation='vertical',color=['b','r'],label=['ml_student','py_student'])
#plt.hist(py_student_age,bins,rwidth=0.8,orientation='vertical',color='y',label='py_student')
plt.legend()
plt.title('ML STUDENT AGE HISTOGRAMS')
plt.xlabel('student agr category')
plt.ylabel('number of studen age')
plt.show()
```
