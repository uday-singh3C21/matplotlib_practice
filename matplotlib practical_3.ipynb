# ploting bar chart

```python
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib import style
import random 
```

```python
classes=['python','R','AI','ML','DS']
class1_Student=np.random.randint(10,30,(5))
class2_Student=np.random.randint(10,30,(5))
class3_Student=np.random.randint(10,30,(5))
print(class1_Student)
print(class2_Student)
print(class3_Student)
```

```python
plt.bar(classes,class1_Student)
```

```python
plt.bar(classes,class2_Student)
```

```python
plt.bar(classes,class3_Student)
```

```python
plt.barh(classes,class1_Student,color='b',edgecolor='r')
```

# how to use multiple bar in the same bar_graph

```python
plt.barh(classes,class1_Student,color='b',edgecolor='r')
plt.title("bar chart for isip")
plt.xlabel("classes")
plt.ylabel("number of sundernts")
plt.show()
```

```python
plt.bar(classes,class1_Student,color='b',edgecolor='r')
plt.bar(classes,class2_Student,color='r',edgecolor='m')
plt.title("bar chart for isip")
plt.xlabel("classes")
plt.ylabel("number of sundernts")
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_Student,width,color='b')
plt.bar(classes_index+width,class2_Student,width,color='r')
plt.title("bar chart for isip")
plt.xlabel("classes")
plt.ylabel("number of sundernts")
plt.show()
```

```python

plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.3
plt.bar(classes_index,class1_Student,width,color='b')
plt.bar(classes_index+width,class2_Student,width,color='r')
plt.bar(classes_index+(2*width),class3_Student,width,color='y')
plt.xticks(classes_index+(2*width),classes)
plt.title("bar chart for isip")
plt.xlabel("classes")
plt.ylabel("number of sundernts")
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.3
plt.barh(classes_index,class1_Student,width,color='b')
plt.barh(classes_index+width,class2_Student,width,color='r')
plt.barh(classes_index+(2*width),class3_Student,width,color='y')
plt.yticks(classes_index+(2*width),classes)
plt.title("bar chart for isip")
plt.ylabel("classes")
plt.xlabel("number of sundernts")
plt.show()
```
