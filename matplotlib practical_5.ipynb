# pie chart 

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
```

```python
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
plt.pie(class1_student)
plt.show()
```

```python
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
plt.pie(class1_student,labels=classes)
plt.show()
```

```python
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
explode=[0.1,0,0.1,0,0]
plt.pie(class1_student,labels=classes,explode=explode)
plt.show()
```

```python
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
explode=[0.1,0,0.1,0,0]
plt.pie(class1_student,labels=classes,explode=explode,autopct="%0.1f%%")
plt.show()
```

```python

classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
explode=[0.01,0.01,0.01,0.01,0.01]
plt.pie(class1_student,labels=classes,explode=explode,autopct="%0.1f%%",frame=True,center=(2,3),radius=1.1)
plt.show()
```

```python
wedgeprops={"linewidth":2,"width":1,"edgecolor":"k"}
classes=["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student=[45,15,35,25,30]
explode=[0.05,0.01,0.1,0.01,0.01]
plt.pie(class1_student,labels=classes,explode=explode,autopct="%0.1f%%",frame=True,center=(2,3),wedgeprops=wedgeprops,radius=1.1)
plt.legend()
plt.show()
```
